```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

imagen = cv.imread('imgpack/Monobanda/2_Corsica.jpg',cv.IMREAD_GRAYSCALE)
```

```python
def histograma(img):
    hist = {i:0 for i in range(0,256)}
    for l in img:
        for x in l:
            hist[x] = hist[x]+1
    plt.bar(hist.keys(),hist.values())
```

```python
histograma(imagen)
```

```python
#Comprobación del histograma con función directa de opencv
plt.hist(imagen.ravel(),256,[0,256]);plt.show()
```
