```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import ipynb.fs.defs.Ejercicio_1 as ej1 #Contiene la funcion histograma
import ipynb.fs.defs.Ejercicio_2 as ej2 #Contiene la funcion expan
import ipynb.fs.defs.Ejercicio_3 as ej3 #Contiene la funcion corte e histograma modificado

imagen = cv.imread('imgpack/Monobanda/2_Corsica.jpg',cv.IMREAD_GRAYSCALE)
```

```python
def ecual(img):
    resimg = img.copy()
    h = ej3.histograma(img)
    FA = np.cumsum(h)
    FE = 255/img.size
    k = 0
    
    for x in range(0,len(FA)):
        FA[x] = np.rint(FA[x]*FE)
    
    
    for i in range(int(np.sqrt(resimg.size))):
        for j in range(int(np.sqrt(resimg.size))):
                resimg[i,j] = FA[resimg[i,j]]
                
    return resimg
```

```python
img_exp = ecual(imagen)

plt.hist(img_exp.ravel(),255,[0,254],width=1);plt.show()

plt.figure(1)
plt.imshow(imagen, cmap='gray')

plt.figure(2)
plt.imshow(img_exp, cmap='gray')
```
