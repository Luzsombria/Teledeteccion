# Ejercicio 4

En este ejercicio emplearemos la funciÃ³n mascara para resaltar detalles en imagenes.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import ipynb.fs.defs.Ejercicio_3 as ej3

imagen = cv.imread('imgpack/Monobanda/2_Corsica.jpg',cv.IMREAD_GRAYSCALE)
peninsula = cv.imread('imgpack/peninsula.png',cv.IMREAD_GRAYSCALE)
```

```python
def calculaMask(img, NDmin, NDmax):
    
    mask = np.zeros([img.shape[0],img.shape[1]], dtype=np.int16)
    
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if NDmax >= img[i,j] >= NDmin:
                mask[i,j] = 1
            else:
                mask[i,j] = 0
    
    return mask
```

```python
mask_aguaYRelieves = calculaMask(imagen,0,70)
msk_img_aguaYRelieves = ej3.mascara(imagen,mask_aguaYRelieves)
plt.imshow(msk_img_aguaYRelieves, cmap='gray')
plt.show()
```

### Como de nuestra imagen tipo no podemos dar demasiados ejemplos, ya que es una imagen con muy pocos detalles donde solo podemos apreciar parte del mar y de la costa en la que aparecen unas montaÃ±as, usaremos otra imagen para comprobar la funciÃ³n anterior

```python
mask_agua = calculaMask(peninsula,0,30)
msk_img_agua = ej3.mascara(peninsula,mask_agua)
plt.imshow(msk_img_agua, cmap='gray')
plt.show()
```

```python
mask_nubesynieve = calculaMask(peninsula,170,255)
msk_img_nubesynieve = ej3.mascara(peninsula,mask_nubesynieve)
plt.imshow(msk_img_nubesynieve, cmap='gray')
plt.show()
```

```python
cv.imwrite("soluciones/ej4-aguaYRelieves.png",msk_img_aguaYRelieves)
cv.imwrite("soluciones/ej4-agua.png",msk_img_agua)
cv.imwrite("soluciones/ej4-nubesynieve.png",msk_img_nubesynieve)
```
