En este ejercicio definiremos la funcion ndvi que generara la imagen correspondiente del indice de vegetacion al operar los valores de las imagenes de banda R y NIR.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

LSR = cv.imread('imgpack/Landsat C/Landsat_C_04-11_R.png')
LSNIR = cv.imread('imgpack/Landsat C/Landsat_C_05-11_NIR.png')
```

```python
def ndvi(path_nir,path_r):
    img_nir = cv.imread(path_nir, cv.IMREAD_GRAYSCALE)
    img_r = cv.imread(path_r, cv.IMREAD_GRAYSCALE)
    ndvi_res = np.zeros((img_r.shape[0],img_r.shape[1]), np.uint8)
    
    for i in range(ndvi_res.shape[0]):
        for j in range(ndvi_res.shape[1]):
            nir = float(img_nir[i,j])
            r = float(img_r[i,j])
            if (nir + r) != 0:
                ndvi_res[i,j] = (((nir - r) / (nir + r)) + 1)*255/2
            else :
                ndvi_res[i,j] = 1
    
    return ndvi_res
```

```python
imgNDVI = ndvi("imgpack/Landsat C/Landsat_C_05-11_NIR.png", "imgpack/Landsat C/Landsat_C_04-11_R.png")
plt.imshow(imgNDVI, cmap = 'gray')
plt.show()
```

```python
cv.imwrite("soluciones/Ej 3 - NVDI.png",imgNDVI)
```
