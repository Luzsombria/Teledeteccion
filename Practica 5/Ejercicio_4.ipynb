# Ejercicio 4

En este ejercicio definiremos la funcion densi que se encargara de realizar un density slicing en imagenes monobanda.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

NDVI = cv.imread('soluciones/Ej 3 - NVDI.png', cv.IMREAD_GRAYSCALE)
```

```python
def densi(img):
    
    densi_res = np.zeros((img.shape[0], img.shape[1], 3), np.uint8)
    
    for i in range(densi_res.shape[0]):
        for j in range(densi_res.shape[1]):
            if 33 < img[i,j] <= 128:
                densi_res[i,j] = [127,127,127]
            elif 128 < img[i,j] <= 136:
                densi_res[i,j] = [255,128,0]
            elif 136 < img[i,j] <= 144:
                densi_res[i,j] = [255,0,128]
            elif 144 < img[i,j] <= 152:
                densi_res[i,j] = [128,255,0]
            elif 152 < img[i,j] <= 160:
                densi_res[i,j] = [0,255,128]
            elif 160 < img[i,j] <= 168:
                densi_res[i,j] = [128,0,255]
            elif 168 < img[i,j] <= 176:
                densi_res[i,j] = [0,128,255]
            elif 176 < img[i,j] <= 184:
                densi_res[i,j] = [247,191,190]
            elif 184 < img[i,j] <= 192:
                densi_res[i,j] = [190,247,191]
                
    
    return densi_res
```

```python
DS = densi(NDVI)
```

```python
plt.imshow(cv.cvtColor(DS, cv.COLOR_BGR2RGB))
plt.show()
```

```python
cv.imwrite("soluciones/Ej 4 - Density scaling.png", DS)
```
