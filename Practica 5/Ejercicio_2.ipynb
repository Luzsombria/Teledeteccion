En este ejercicio definiremos la funci√≥n seudo, que permite modificar una imagen monobanda a una imagen multicolor para determinados valores.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

imagen = cv.imread('imgpack/Monobanda/2_Corsica.jpg', cv.IMREAD_GRAYSCALE)
```

```python
def seudo(img, corte):
    
    resimg = np.zeros((img.shape[0],img.shape[1],3), np.uint8)
    
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i,j] < corte:
                resimg[i,j] = [0,0,255]
            elif img[i,j] >= corte:
                resimg[i,j] = [255,0,0]
    
    return resimg
```

```python
bicolor = seudo(imagen, 128)
plt.imshow(bicolor)
plt.show()
```

```python
M = seudo(imagen,40)
MC = seudo(imagen,150)
```

```python
plt.figure(figsize=(20, 20))
plt.subplot(2, 2, 1)
plt.imshow(imagen, cmap = 'gray')
plt.title("Original", fontsize = 30)

plt.subplot(2, 2, 2)
plt.imshow(bicolor)
plt.title("Seudo 128", fontsize = 30)

plt.subplot(2, 2, 3)
plt.imshow(M)
plt.title("Seudo Mar", fontsize = 30)

plt.subplot(2, 2, 4)
plt.imshow(MC)
plt.title("Seudo Monta√±as", fontsize = 30)

plt.tight_layout()
plt.show()
```

Debido a que en la imagen usada no se pueden apreciar demasiados detalles usaremos la misma imagen que en el documento.

```python
nocturna = cv.imread('imgpack/nocturna.png', cv.IMREAD_GRAYSCALE)
seudoNocturna = seudo(nocturna, 128)
nubes = seudo(nocturna, 90)
mar = seudo(nocturna, 20)
```

```python
plt.figure(figsize=(20, 20))
plt.subplot(2, 2, 1)
plt.imshow(nocturna, cmap = 'gray')
plt.title("Original", fontsize = 30)

plt.subplot(2, 2, 2)
plt.imshow(seudoNocturna)
plt.title("Seudo 128", fontsize = 30)

plt.subplot(2, 2, 3)
plt.imshow(nubes)
plt.title("Seudo Nubes", fontsize = 30)

plt.subplot(2, 2, 4)
plt.imshow(mar)
plt.title("Seudo Mar", fontsize = 30)

plt.tight_layout()
plt.show()
```

```python
cv.imwrite("soluciones/Ej 2 - Seudo 128 Corsica.png", cv.cvtColor(bicolor, cv.COLOR_BGR2RGB))
cv.imwrite("soluciones/Ej 2 - Seudo Mar Corsica.png", cv.cvtColor(M, cv.COLOR_BGR2RGB))
cv.imwrite("soluciones/Ej 2 - Seudo Monta√±as.png", cv.cvtColor(MC, cv.COLOR_BGR2RGB))
cv.imwrite("soluciones/Ej 2 - Seudo 128 Nocturna.png", cv.cvtColor(seudoNocturna, cv.COLOR_BGR2RGB))
cv.imwrite("soluciones/Ej 2 - Seudo Nubes.png", cv.cvtColor(nubes, cv.COLOR_BGR2RGB))
cv.imwrite("soluciones/Ej 2 - Seudo Mar Nocturna.png", cv.cvtColor(mar, cv.COLOR_BGR2RGB))
```
