```python
import numpy as np

x = np.arange(0,10,1)
print(x)

y = x ** 2
print(y)

z = x*2
print(z)

c = np.arange(10,21,2)
print(c)

d = c-1
print(d)
```

```python
print(np.sin(np.pi/4))

x = np.arange(0,np.pi,0.01)
print(x)

y = np.sin(x)
print(y[0], "...", y[314])

z = np.cos(x)
print(z[0], "...", z[314])
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

plt.figure(1)
plt.plot(x)
plt.ylabel('Lista de valores')
plt.show()
```

```python
plt.figure(2)
plt.plot(y)
plt.ylabel('Lista de valores y')
```

```python
plt.figure(3)
plt.plot(x,y)
plt.ylabel('Sin(x)')
plt.xlabel('x')
```

```python
plt.figure(4)
plt.plot(x,y,'r')
plt.plot(x,z,'g')
plt.legend(['Sin(x)', 'Cos(x)'])
plt.xlabel('x')
plt.ylabel('Values')
```
