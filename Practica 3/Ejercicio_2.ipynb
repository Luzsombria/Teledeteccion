En este ejercicio simplemente acudiremos a la funci√≥n definida en el ejercicio 1 y le aplicaremos a una imagen monobanda diferentes tipos de filtro.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
# Para importar las funciones de otro notebook tenemos que ejecutar primero los siguientes comandos:
# pip install ipynb
# import ipynb.fs.defs.nombre_del_notebook_a_importar as nombre
import ipynb.fs.defs.Ejercicio_1 as ej1

imagen = cv.imread('imgpack/Monobanda/2_Corsica.jpg',cv.IMREAD_GRAYSCALE)
```

```python
media = np.array([[1,1,1],[1,1,1],[1,1,1]])
medImg = np.array([[1,1,1],[1,2,1],[1,1,1]])
gauss = np.array([[1,2,1],[2,4,2],[1,2,1]])
ejemplo = np.array([[12,14,17,24,32,30],
                    [19,18,17,26,36,40],
                    [26,21,17,28,40,42],
                    [20,19,18,26,33,35],
                    [14,17,20,23,27,20],
                    [35,14,22,13,40,21]])
```

```python
print(ej1.filtro(ejemplo,media))
```

```python
print(ej1.filtro(ejemplo,medImg))
```

```python
print(ej1.filtro(ejemplo,gauss))
```

```python
imgMed = ej1.filtro(imagen,media)
```

```python
imgMedImg = ej1.filtro(imagen,medImg)
```

```python
imgGauss = ej1.filtro(imagen,gauss)
```

```python
fig, axs = plt.subplots(2,2)

axs[0,0].imshow(imagen, cmap='gray', aspect='auto')
axs[0,1].imshow(imgMed, cmap='gray', aspect='auto')
axs[1,0].imshow(imgMedImg, cmap='gray', aspect='auto')
axs[1,1].imshow(imgGauss, cmap='gray', aspect='auto')
```

```python
cv.imwrite("soluciones/EJ2_Matriz_Media.png", imgMed)
cv.imwrite("soluciones/EJ2_Matriz_MediaImagen.png", imgMedImg)
cv.imwrite("soluciones/EJ2_Matriz_Gaussiana.png", imgGauss)
```
